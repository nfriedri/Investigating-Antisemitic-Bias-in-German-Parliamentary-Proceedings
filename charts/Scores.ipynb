```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
from google.colab import files
import numpy as np

def plot_bias_line(data, statistic, title):
  data_list = []
  for d in data:
    if d["measure"] == statistic:
       d["view"] = d["specification"].split("_")[2] if len(d["specification"].split("_")) > 2 else d["specification"].split("_")[1]
       data_list.append(d)
  # print(data_list)
  df = pd.DataFrame(data_list)
  sns.set_theme()

  g = sns.lineplot(data=df,
      x="slice", 
      y="score", 
      hue="view", 
      style="view", 
      alpha=.6, 
      palette="hls"
  )


  #g.despine(left=True)
  #g.set_axis_labels("", statistic)
  #g.legend.set_title("")

  plt.savefig(title + ".pdf")
  files.download(title + ".pdf")
```
